```python
names=["Avik","Raj","Kamal"]
marks=[9.8,8.8,9.7]
plt.bar(names,marks,color=['red','green','blue'],width=.1)
plt.xlabel("Names")
plt.ylabel("Marks")
plt.title("Names vs Marks")
plt.show()
```

```python
goal_types=["penalty","field","Free kick"]
goals=[22,31,12]
plt.pie(goals,labels=goal_types,autopct="%.2f%%",explode=(0.1,0.1,0.1),shadow=True)
plt.title("Goals Scored by Messi")
plt.show()
```
